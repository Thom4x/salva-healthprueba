# Consumo de datos desde Azure Storage

Como parte de la prueba técnica, el conjunto de datos preparado para el modelado fue publicado en Azure Blob Storage.

El objetivo de esta etapa es demostrar un flujo moderno de acceso a datos, donde el dataset se consume directamente desde un servicio de almacenamiento en la nube en lugar de ser leído desde el disco local.

Para este proyecto se utilizó **Azure Blob Storage**, debido a su integración con el ecosistema de Azure y su uso frecuente en proyectos de ingeniería de datos y Machine Learning.

```python
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
# Cargar variables de entorno
load_dotenv()
```

```python
AZURE_BLOB_URL = os.getenv("AZURE_BLOB_URL")

if AZURE_BLOB_URL is None:
    raise ValueError(
        "La variable AZURE_BLOB_URL no fue encontrada en el archivo .env."
    )

df_cloud = pd.read_csv(AZURE_BLOB_URL)

df_cloud.head()
```

```python
df_cloud.info()
```

El dataset fue consumido correctamente desde Azure Blob Storage mediante una lectura remota utilizando `pandas.read_csv()`.

Este flujo elimina la dependencia del almacenamiento local y permite acceder a los datos desde un servicio en la nube
