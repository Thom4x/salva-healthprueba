# Ingeniería de características

En esta etapa se proponen algunas variables derivadas que podrían aportar información adicional al modelo de clasificación.

El objetivo no es mejorar el baseline desarrollado en esta prueba, sino identificar posibles características que puedan incrementar el poder predictivo en futuras iteraciones. Las variables propuestas se construyen a partir de la información disponible, manteniendo un criterio clínico y de interpretabilidad.

```python
import pandas as pd

df = pd.read_csv("../data/model-data/pacientes_model.csv")

df.head()
```

## Feature 1 - Índice de Masa Corporal (IMC)

El IMC es una medida ampliamente utilizada para relacionar el peso y la estatura de una persona.

Se calcula como:

\[
IMC = \frac{peso}{altura^2}
\]

Esta variable podría aportar información relevante, ya que el sobrepeso y la obesidad están asociados con diversas enfermedades cardiovasculares.

```python
df["imc"] = df["peso_kg"] / ((df["altura_cm"] / 100) ** 2)

df[["peso_kg", "altura_cm", "imc"]].head()
```

## Feature 2 - Grupo de edad

Se agruparon las edades en categorías con el fin de capturar posibles diferencias fisiológicas entre distintos grupos poblacionales.

Esta representación puede facilitar el aprendizaje de patrones que no sean completamente lineales respecto a la edad.

```python
bins = [0, 40, 60, 120]

labels = [
    "Joven",
    "Adulto",
    "Adulto mayor"
]

df["grupo_edad"] = pd.cut(
    df["edad_paciente"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

df["grupo_edad"].value_counts()
```

## Feature 3 - Categoría de frecuencia cardíaca

La frecuencia cardíaca media se agrupó en categorías clínicas generales para facilitar su interpretación.

Esta transformación podría permitir que modelos basados en árboles identifiquen patrones asociados con rangos fisiológicos específicos.

```python
bins = [0, 60, 100, 250]

labels = [
    "Baja",
    "Normal",
    "Alta"
]

df["categoria_fc"] = pd.cut(
    df["frecuencia_cardiaca_media_bpm"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

df["categoria_fc"].value_counts()
```

```python
# Ver el DataFrame con las mejoras propuestas
df.head()
```

Se propusieron tres variables derivadas construidas a partir de la información disponible en el conjunto de datos:

- **IMC**, como indicador del estado nutricional.
- **Grupo de edad**, para representar diferencias entre rangos etarios.
- **Categoría de frecuencia cardíaca**, basada en intervalos fisiológicos generales.

Estas características no fueron incorporadas al modelo baseline con el fin de mantener una comparación sencilla e interpretable. Sin embargo, constituyen hipótesis de trabajo que podrían evaluarse en futuras iteraciones para analizar su impacto sobre el desempeño del modelo.
